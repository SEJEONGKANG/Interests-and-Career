```python
import requests
import json
```

```python
header = {
    'user-agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 11_0 like Mac OS X) AppleWebKit/604.1.38 (KHTML, like Gecko) Version/11.0 Mobile/15A372 Safari/604.1',
    'x-requested-with': 'XMLHttpRequest'
}
```

```python
s = requests.Session()
s.headers.update(header)
```

```python
def SEARCH_DISTANCE_URL(start, end):
    return 'https://m.map.naver.com/spirra/findCarRoute.nhn?route=route3&output=json&result=web3&coord_type=latlng&search=2&car=0&mileage=12.4' + '&start={}&destination={}'.format(start, end)
```

```python
def GET_INFO(start, end):
    res = s.get(SEARCH_DISTANCE_URL(start, end)).text
    res_dict = json.loads(res)
    
    target = res_dict['routes'][0]['summary']
    distance = target['distance']
    sec = target['duration']
    taxi_fare = target['taxi_fare']
    #print('검색완료')
    #print('출발지: {}, 도착지: {}'.format(start, end))
    #print('추천경로 이동 거리: {}km'.format(distance/1000))
    #print('예상 소요시간: {}분'.format(sec/60))
    #print('예상 택시요금: {}원'.format(taxi_fare))
    print('출발지: {}, 도착지: {}, 추천경로 이동 거리: {}km'.format(start, end, distance/1000))
    print('예상 소요시간: {}초, 속도: {}m/s'.format(sec, distance/sec))
    return sec, distance/sec
```

```python
node_dict = {}
for i in range(len(node_data["NODE_ID"])):
    node_dict[node_data["NODE_ID"][i]] = [node_data["위도"][i], node_data["경도"][i]]
```

```python
link_dict = {}
for i in range(len(link_data["LINK_ID"])):
    ID = link_data["LINK_ID"][i]
    link_dict[ID] = node_dict[link_data["F_NODE"][i]] + node_dict[link_data["T_NODE"][i]] + [link_data["LENGTH"][i]]
```

```python
import time
import random
```

```python
import datetime
```

```python
time_lst = []
speed_lst = []

c = datetime.datetime.now()
print(c)
for key in link_dict.keys():
    while True:
        try:
            x1, y1, x2, y2, d = link_dict[key]
            start = str(y1) + "," + str(x1)
            end = str(y2) + "," + str(x2)
            t, speed = GET_INFO(start,end)
            time_lst.append(t)
            speed_lst.append(speed)
            time.sleep(1+random.random()*2)
            break
        except:
            print(datetime.datetime.now())
            time.sleep(60)
print(c)
print(datetime.datetime.now())
```
