```python
result_df = pd.read_csv(f"{ROOT}/data/AHP_result3.csv")
#result_df = result_df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df = result_df.pivot_table('점수', index='강좌명', columns='프로그램명')

# User-based 추천을 위해 데이터프레임을 전치하여 사용자 기반 유사도를 계산
df_transposed = df

# 유사도 계산을 위해 데이터를 스케일링
df_scaled = df_transposed.apply(lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x)), axis=1)

# NaN 값을 0으로 채움
df_scaled = df_scaled.fillna(0)

# 사용자 간 유사도 계산
user_sim = cosine_similarity(df_scaled, df_scaled)

# 사용자 간 유사도를 데이터프레임으로 변환
user_sim = pd.DataFrame(user_sim, index=df_scaled.index, columns=df_scaled.index)

print(user_sim.shape)
user_sim
```
