```python
protectzone = station_map
def drawmap():
    guromap = folium.Map(location=[37.51651696477263,127.03873607397672], zoom_start=13,
                        tiles = "CartoDB positron")
    
    # 강남구 경계선 테두리와 내부 색칠하기
#     guromap.choropleth(geo_data = "dataset\guroline.json", 
#                     fill_color = "#FFFFFF",
#                     fill_opacity = 0.2,
#                    line_color = "#0491FD",line_weight = 1.5)
    
    # 스쿨존 표시하기
    for i in range(protectzone.shape[0]):
        protect_a = protectzone.위도.values[i]
        protect_b = protectzone.경도.values[i]
        folium.CircleMarker([protect_a,protect_b],
                            color="#F6B316", fill="#F9CC02",
                            weight = 2,
                            radius = 10, opacity = 1).add_to(guromap)
    
    return guromap
```
